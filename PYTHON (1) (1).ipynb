```python
# Python Operators Demonstration

a = 10
b = 4
c = 10
d = 20

print("\n--- Arithmetic Operators ---")
print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b}") # Division returns a float
print(f"a % b = {a % b}") # Modulus
print(f"a ** b = {a ** b}") # Exponentiation
print(f"a // b = {a // b}") # Floor division

print("\n--- Comparison Operators ---")
print(f"a == c (10 == 10) is {a == c}")
print(f"a != b (10 != 4) is {a != b}")
print(f"a > b (10 > 4) is {a > b}")
print(f"a < b (10 < 4) is {a < b}")
print(f"a >= c (10 >= 10) is {a >= c}")
print(f"a <= b (10 <= 4) is {a <= b}")

print("\n--- Logical Operators ---")
x = True
y = False
print(f"x and y (True and False) is {x and y}")
print(f"x or y (True or False) is {x or y}")
print(f"not x (not True) is {not x}")

print("\n--- Assignment Operators ---")
e = 5
print(f"e = {e}")
e += 3 # e = e + 3
print(f"e += 3  => e = {e}")
e *= 2 # e = e * 2
print(f"e *= 2  => e = {e}")
# ... many more like -=, /=, %=, //=, **=

print("\n--- Identity Operators ---")
# 'is' checks if two variables refer to the same object in memory
print(f"a is c (both are 10) is {a is c}") # Should be True for small integers due to optimization
list1 = [1, 2]
list2 = [1, 2]
list3 = list1
print(f"list1 is list2 is {list1 is list2}")
print(f"list1 is list3 is {list1 is list3}")

print("\n--- Membership Operators ---")
# 'in' checks if a value is present in a sequence (string, list, tuple, set, dictionary keys)
my_list = [10, 20, 30, 40]
print(f"10 in my_list is {10 in my_list}")
print(f"50 not in my_list is {50 not in my_list}")
my_string = "Hello World"
print(f"'World' in my_string is {'World' in my_string}")

# Note: Bitwise operators (&, |, ^, ~, <<, >>) are also available but less commonly used in general programming.
```

### **High Order Function->follow atleast one condition**
1. Take one or more ftn as arguments.
2. Returns a ftn as its result.

```python
def transform(f,L):
  output=[]
  for i in L:
    output.append(f(i))
  print(output) # Corrected from 'Output' to 'output'
L=[1,2,3,4,5]
transform(lambda x:x**2,L)
```

```python
def return_sum(func,L):
  result = 0 # Initialize result
  for i in L:
    if func(i):
      result +=i
  return result
L=[11,23,55,7,2,0,98,67,5,3,9,8]

print(return_sum(lambda x:x%2==0,L))
print(return_sum(lambda y:y%2!=0,L))
print(return_sum(lambda z:z%3==0,L)) # Added missing parenthesis
```

```python
#Multi level inheritance
class car:
  color="Read"
  @staticmethod
  def start():
    print("Start the car")
  @staticmethod
  def Stop():
    print("Stop the car")

class Toyota_Car(car):
  def __init__(self,brand):
    self.brand=brand

class Honda_Car(Toyota_Car):
  def __init__(self, brand, type): # Added brand parameter
    super().__init__(brand) # Super Constructor
    super().start() # Super Method
    self.type=type

c1=Toyota_Car("Fortuner")
c2=Honda_Car("Honda", "Civic") # Pass both brand and type
print(c2.brand,c2.type,c2.color)
c2.start()
c2.Stop()
```

Super Method uswd to access methods of the parent class.

```python
#Multiple inheritance
class A:
  varA="Welcom A"
class B:
  varB="Welcom B"
class C(A,B):
  pass

C1=C()
print(C1.varA,C1.varB)
```

Property Decorator

 we use @property decorator on any method in the class to use the method as a property

```python
#Attriute value depend on ftn, then use ftn as a property
class student:
  def __init__(self,phy,che,math):
    self.phy=phy
    self.che=che
    self.math=math

  @property
  def avg(self):
    return (self.phy+self.che+self.math)/3

stu1=student(23,45,67)
print(stu1.avg)
stu1.phy=89
print(stu1.avg)
```

```python
class MyComplexNumber:
  def __init__(self,r,i):
    self.real=r
    self.img=i

  def __add__(self,c):
    return MyComplexNumber(self.real+c.real,self.img+c.img)

num1=MyComplexNumber(5,6)
num2=MyComplexNumber(6,7)
sum_result=num1+num2 # Changed 'sum' to 'sum_result' to avoid shadowing built-in sum function
print(sum_result.real,sum_result.img)
```
