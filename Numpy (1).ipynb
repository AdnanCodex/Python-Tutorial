What is Numpy?
</div>

NumPy = Numerical Python

🤞A Python library for scientific computing, designed for fast numerical calculations.

🤞Handles multi-dimensional arrays and matrices efficiently.

🤞Used in Machine Learning, Data Science, Image Processing, and scientific simulations.

<div style="font-size: 25px; font-weight: bold; color: #33adff; text-align: left; margin-top: 20px;">
Why Numpy is needed
</div>

Python lists were slow for large data and scientific computations.

Lists are heterogeneous and dynamic, making math operations inefficient.

NumPy arrays (ndarray) are homogeneous and stored in continuous memory, making them 50x faster.

Multi-dimensional arrays are easier to handle with NumPy.

Large data operations, matrix multiplications, and ML algorithms require NumPy.


Summary:
NumPy was created because Python lists were inefficient for numerical and scientific computations.

<div style="font-size: 25px; font-weight: bold; color: #33adff; text-align: left; margin-top: 20px;">
How Numpy is built
</div>

Written in C language for speed.

Python acts as an interface, while C handles heavy computations.


In short:

> NumPy = Python + C (fast backend).

```python
# Scalar Operation
import numpy as np
a1=np.array([1,2,3,4])
print(a1*2, a1/2, a1|2, a1**2,sep="\n") #Arthimatic Operator
print( a1 >1,a1!=2, a1 & 2,2 in a1, sep="\n") #Relational Operator (bitwise AND)
```

```python
#AND
print((a1 > 1) & (a1 < 4))
#OR
print((a1 == 1) | (a1 == 4))
# Element-wise NOT (~)
print(~(a1 > 2))
# Combining conditions to select elements
print(a1[(a1 > 1) & (a1 < 4)])
```

```python
#Vector Operation
a1=np.array([1,2,3,4])
a2=np.array([[1,2,3,4],[5,6,7,8]])
a3=np.array([[1,2,3,4],[5,6,7,8]])
print(a1+a2,sep="\n")
```

```python
#Sigmoid S(x)=1/(1+e^-x)
def sigmoid(array):
  return 1/(1+np.exp(-(array)))

a=np.arange(10)
print(sigmoid(a))
```

```python
#Mean Square Error MSE=1/n Semmation i=1 to n (Yi-Y'i)^2
#Use in linear regration-> Prediction->calculate error then take summation
actual=np.random.randint(1,50,25)
predicted=np.random.randint(1,50,25)

def mse(actual,predicted):
  return np.mean((actual-predicted)**2)

print(mse(actual,predicted))
```
